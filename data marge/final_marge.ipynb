```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',1000)
pd.set_option('display.max_rows',1000)
```

```python
game_df = pd.read_csv("../data clean/data/game_schedule_clean.csv")
batting_team_df = pd.read_csv("../data clean/data/batting_team_clean.csv")
batting_player_df = pd.read_csv("../data clean/data/batting_player_clean.csv")
pitchers_team_df = pd.read_csv("../data clean/data/pitching_team_clean.csv")
pitchers_player_df = pd.read_csv("../data clean/data/pitchers_player_clean.csv")
```

```python
print("Game Data :- ",game_df.shape)
print("Batting Team Data :- ",batting_team_df.shape)
print("Batting Player Data :- ",batting_player_df.shape)
print("Pitchers Team Data :- ",pitchers_team_df.shape)
print("Pitchers Player Data :- ",pitchers_player_df.shape)
```

```python
datasets = {
    "game_schedule": game_df,
    "batting_team": batting_team_df,
    "batting_player": batting_player_df,
    "pitching_team": pitchers_team_df,
    "pitching_player": pitchers_player_df
}

for name, df in datasets.items():
    df_clean = df.drop_duplicates(subset=['gamePk'], keep='first')
    print(f"{name}: cleaned rows = {len(df_clean)}")
    
    # overwrite original dataframe
    datasets[name] = df_clean

# Ab datasets me sab clean files ready hain
game_df = datasets["game_schedule"]
batting_team_df = datasets["batting_team"]
batting_player_df = datasets["batting_player"]
pitchers_team_df = datasets["pitching_team"]
pitchers_player_df = datasets["pitching_player"]
```

```python
# Start with Game Schedule
df = game_df.copy()

# Merge team-level and player-level features
df = df.merge(batting_team_df, on='gamePk',how="inner")
df = df.merge(batting_player_df, on='gamePk', how='left')
df = df.merge(pitchers_team_df, on='gamePk', how='left')
df = df.merge(pitchers_player_df, on='gamePk', how='left')
```

```python
df.drop(columns = ['season', 'gamePk', 'away_team_name_x', 'home_team_name_x', 'date', 'home_team_id_x', 'home_team_name_y',
                   'away_team_id_x', 'away_team_name_y', 'home_team_id_y', 'away_team_id_y', 'home_team_id', 'away_team_id',
                   'home_pitching_player_K_BB_ratio', 'home_pitching_player_ground_to_fly_ratio', 'away_pitching_player_K_BB_ratio',
                   'away_pitching_player_ground_to_fly_ratio'],inplace = True)
```

```python
df['target'] = (df['home_team_wins'] > df['away_team_wins']).astype(int)
```

```python
df.to_csv("data/final_marge_data.csv", index = False)
```

```python
from sklearn.preprocessing import LabelEncoder

# Columns to encode (categorical)
categorical_cols = ['dayNight', 'venue_name', 'home_team_name', 'away_team_name']  

# Initialize LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# Check encoded columns
print(df[categorical_cols].head())
```

```python
df.to_csv("data/final_data.csv", index = False)
```
