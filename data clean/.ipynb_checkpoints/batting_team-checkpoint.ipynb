```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',1000)
pd.set_option('display.max_rows',1000)
```

```python
df = pd.read_csv("../data fetch/data/batting_team_2024.csv")
df.shape
```

```python
# Batting average last 5 games
df['avg_last_5'] = df.groupby('team_id')['avg'].transform(lambda x: x.rolling(5, min_periods=1).mean().shift(1))

# OPS last 5 games
df['ops_last_5'] = df.groupby('team_id')['ops'].transform(lambda x: x.rolling(5, min_periods=1).mean().shift(1))

# Home runs rate last 5 games
df['hr_rate_last_5'] = df.groupby('team_id')['homeRuns'].transform(lambda x: x.rolling(5, min_periods=1).mean().shift(1))

# Strikeouts rate last 5 games
df['strikeout_rate_last_5'] = df.groupby('team_id')['strikeOuts'].transform(lambda x: x.rolling(5, min_periods=1).mean().shift(1))

# Walk rate last 5 games
df['walk_rate_last_5'] = df.groupby('team_id')['baseOnBalls'].transform(lambda x: x.rolling(5, min_periods=1).mean().shift(1))
```

```python
# Stolen base success rate last 5 games 

def stolen_success(x):
    sb_sum = x['stolenBases'].rolling(5, min_periods=1).sum()
    cs_sum = x['caughtStealing'].rolling(5, min_periods=1).sum()
    rate = sb_sum / (sb_sum + cs_sum)
    return rate.shift(1) # shift ensures pre-game 

df['stolen_success_last_5'] = df.groupby('team_id', group_keys=False).apply(stolen_success).reset_index(level=0, drop=True)
```

```python
df['avg_last_5'] = df['avg_last_5'].fillna(df['avg'].mean())
df['ops_last_5'] = df['ops_last_5'].fillna(df['ops'].mean())
df['hr_rate_last_5'] = df['hr_rate_last_5'].fillna(0)
df['strikeout_rate_last_5'] = df['strikeout_rate_last_5'].fillna(0)
df['walk_rate_last_5'] = df['walk_rate_last_5'].fillna(0)
df['stolen_success_last_5'] = df['stolen_success_last_5'].fillna(0.5)
```

```python
df.drop(columns = ['atBats', 'hits','doubles', 'triples', 'homeRuns', 'rbi', 'strikeOuts',
                   'baseOnBalls', 'avg', 'obp', 'slg', 'ops', 'stolenBases', 'caughtStealing',
                   'plateAppearances', 'leftOnBase'], inplace = True)
```

```python
df.to_csv("data/team_data.csv", index = False)
```

```python
# Home team ka data
batting_home = (
    df[df['team_side'] == 'home']          # home rows filter
    .drop(columns=['team_side'])           # team_side column drop
    .add_prefix('home_')                   # sab column names ke aage home_ lagao
    .rename(columns={'home_gamePk': 'gamePk'})  # gamePk ka naam wapas original karna
)

# Away team ka data
batting_away = (
    df[df['team_side'] == 'away']          # away rows filter
    .drop(columns=['team_side'])
    .add_prefix('away_')
    .rename(columns={'away_gamePk': 'gamePk'})
)

# Merge home & away
df = pd.merge(batting_home, batting_away, on='gamePk', how='inner')
```
