```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',1000)
pd.set_option('display.max_rows',1000)
```

```python
df = pd.read_csv("../data fetch/data/game_schedule_2024.csv")
df.shape
```

```python
df['date'] = pd.to_datetime(df['officialDate'], errors='coerce')
df['date'] = df['date'].sort_values().reset_index(drop=True)
```

```python
df = df[~df['status'].isin(['Postponed', 'Cancelled'])].reset_index(drop=True)
```

```python
df['is_last_game_of_series'] = (df['seriesGameNumber'] == df['gamesInSeries']).astype(int)
df['series_progress'] = df['seriesGameNumber'] / df['gamesInSeries']
```

```python
# Regular MLB Stadiums list
regular_stadiums = [
    'Guaranteed Rate Field', 'Fenway Park', 'Oriole Park at Camden Yards', 'Rogers Centre',
    'Great American Ball Park', 'loanDepot park', 'Kauffman Stadium', 'Minute Maid Park',
    'Globe Life Field', 'Chase Field', 'Tropicana Field', 'American Family Field',
    'Nationals Park', 'T-Mobile Park', 'Wrigley Field', 'Busch Stadium', 'Yankee Stadium',
    'Target Field', 'Oakland Coliseum', 'Truist Park', 'Oracle Park', 'Angel Stadium',
    'PNC Park', 'Citizens Bank Park', 'Citi Field', 'Petco Park', 'Dodger Stadium',
    'Progressive Field', 'Comerica Park', 'Coors Field'
]

# Replace all non-regular stadiums with 'Other'
df['venue_name'] = df['venue_name'].apply(lambda x: x if x in regular_stadiums else 'Other')
```

```python
df.drop(columns = ["gameGuid", "gameDate", "officialDate", "description", "seriesDescription", "isTie",
                   "venue_id", "gameType", "status", "scheduledInnings", "home_score", "away_score",
                   "home_isWinner", "away_isWinner", "away_team_id", "home_team_id"],inplace = True)
```

```python
# df['gamePk'].value_counts().loc[lambda x: x > 1]
# duplicates_df = df[df.duplicated(subset=['gamePk'], keep=False)]

df = df.drop_duplicates(subset=['gamePk'], keep='first')
```
